```python
import pandas as pd
from datetime import datetime
from dateutil.parser import parse
import numpy as np
raw_data = {'employee_name': ['Andy', 'Beth', 'Cindy', "Dale",'Andy', 'Beth', 'Cindy'],
            'employee_id': [123456,789456,654123,963852,1243456,7849456,6534123],
            'date_joined': ['2015-02-15', np.nan, '2017-05-16', "2018-01-15",'2015-02-15', np.nan, '2017-05-16'],
            'age': [45,36,34,25,45,36,34],
            'yrs_of_experience': [24,14,14,4,24,14,14]}
df = pd.DataFrame(raw_data, columns = ['employee_name', 'employee_id', 'date_joined','age', 'yrs_of_experience'])
df
```

```python
#fill na with Dec 1, 2009 as given 
df.date_joined = df.date_joined.fillna('2009-12-01')
```

```python
df
```

```python
df['date_joined_month'] = df['date_joined'].str[:7]
df = df[['date_joined_month', 'employee_id']]
```

```python
df_pivot = df.groupby(['date_joined_month']).employee_id.nunique().reset_index()
```

```python
df_pivot.columns = ['month_joined', 'num_employees']
```
