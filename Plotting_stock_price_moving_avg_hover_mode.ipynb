```python
import pandas as pd
```

```python
aapl = pd.read_csv('data/AAPL.csv')#, parse_dates=['Date'])
```

```python
aapl.head(5)
```

```python
aapl['CloseWA'] = aapl['Close'].rolling(window=7).mean()
```

```python
#plotting the chart
import matplotlib.pyplot as plt
%matplotlib inline

plot = aapl.plot(x='Date', y=['Close', 'CloseWA'],
                 kind='line', title='Plot of stock price')

#change legend
plt.legend(loc='lower left')
```

```python

import plotly.express as px

fig = px.line(aapl, x="Date", y=['Close', 'CloseWA'], title="Hover Stock plotting")
fig.update_traces(mode="markers+lines", hovertemplate=None)
fig.update_layout(hovermode="x")

fig.show()
```
